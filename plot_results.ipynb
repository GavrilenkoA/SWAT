```python
# the results will be as follow
results = pd.read_csv('results/BLAT_ECOLX_Ostermeier2014_esm2_150M_layer_30_compressed.csv', index_col=0)
results
```

```python
results.sort_values('Compression_methd', inplace=True)
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.barplot(data=results, x='Compression_methd', y='R2_score_test', ax=ax[0])
ax[0].set_title('BLAT ECOLX Ostermeier2014')
ax[0].set_ylim(0, 1)
ax[0].set_ylabel('R2 Score')
ax[0].set_xlabel('Compression methd')
sns.barplot(data=results, x='Compression_methd', y='rho_score_test', ax=ax[1])
ax[1].set_title('BLAT ECOLX Ostermeier2014')
ax[1].set_ylim(0, 1)
ax[1].set_ylabel('Rho Score')
ax[1].set_xlabel('Compression methd')
plt.show()
```

```python
def plot_results(res_path):
    results = pd.read_csv(res_path, index_col=0)
    results.sort_values('Compression_methd', inplace=True)
    title = ' '.join(res_path.split('/')[-1].split('_')[:-5])
    plt.figure(figsize=(12, 5))
    sns.barplot(data=results, x='Compression_methd', y='R2_score_test')
    plt.title(title)
    plt.ylim(0, 1)
    plt.ylabel('R2 Score')
    plt.xlabel('Compression methd')
    plt.show()
```

```python
for file in os.listdir('results/regression/'):
    file_path = os.path.join('results/regression/', file)

    plot_results(file_path)
```
